```python
import operator
import numpy as np
import os
from csv import DictReader

datoteke = os.listdir("podatki/")
podatki_o_vozilih = []
for dat in datoteke:
    if "Podatki" in dat and "2020" not in dat:
        neki = DictReader(open('podatki/' + dat, 'rt', encoding='ANSI'), delimiter=';')
        for i in neki:
            if i["P.1.2-Nazivna moc"] != "" and i["P.1.2-Nazivna moc"] != None and i["P.1.2-Nazivna moc"] != "Bencin":
                
                podatki_o_vozilih.append([float(str(i["P.1.2-Nazivna moc"]).replace(",", ".")), i["D.3-Komerc. oznaka"], i["D.1-Znamka"]])
                
podatki_o_vozilih = np.array(podatki_o_vozilih)
```

```python
print(podatki_o_vozilih)
```

Najmočnejše registrirano vozilo.

```python
max = 0
for i in podatki_o_vozilih:
    if float(i[0]) > max:
        max = float(i[0])
        max_model = i[1]
        max_znamka = i[2]
```

```python
print("Najmočnejše novo registrirano vozilo v Sloveniji je " + max_model + ", znamke " + max_znamka + ", ki ima " + str(max) + " kW nazivne moči motorja.")
```

Ker je očitno napaka v podatkih, omejimo najvišjo moč motorja.

```python
max = 0
for i in podatki_o_vozilih:
    if float(i[0]) > max and float(i[0]) < 750:
        max = float(i[0])
        max_model = i[1]
        max_znamka = i[2]
```

```python
print("Najmočnejše novo registrirano vozilo v Sloveniji je " + max_model + ", znamke " + max_znamka + ", ki ima " + str(max) + " kW nazivne moči motorja.")
```
